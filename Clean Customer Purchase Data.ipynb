```python
import pandas as pd
from IPython.display import display

data = {
    'Customer ID': [101, 102, 103, 104, None, 106],
    'Name': ['Alice Smith', 'BOB JONES', 'Charlie', ' Dana White ', 'Emily ', 'Frank'],
    'Email': ['ALICE@EMAIL.COM', 'bob@email.com', None, 'dana@white.com', 'emily@email.com ', 'frank@email.com'],
    'Purchase Amount': ['$250.00', '$300.00', 'Invalid', '$150.75', '$200', ''],
    'Date': ['2023-01-15', '2023/02/20', '15-03-2023', 'not available', '2023.05.10', '2023-06-01']
}

df = pd.DataFrame(data)
df
```

```python
# Strip white spaces from column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Clean 'name' and 'email'
df['name'] = df['name'].str.strip().str.title()
df['email'] = df['email'].str.strip().str.lower()

# Convert 'purchase amount' to numeric
df['purchase_amout'] = df['purchase_amount'].replace('[\$,]', '', regex=True)
df['purchase_amout'] = pd.to_numeric(df['purchase_amount'], errors='coerce')

# Convert 'date' to 'date_time'
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Drop rows with missing customer_id or email
df = df.dropna(subset=['customer_id', 'email'])

# Reset index
df.reset_index(drop=True, inplace=True)

df
```

```python
# Total spend per customer 
total_spend = df.groupby('customer_id')['purchase_amount'].sum()

# Average purchase per customer
avg_spend = df.groupby('customer_id')['purchase_amout'].mean()

print("Total Spend per Customer:\n", total_spend))
print("\nAverage Spend per Customer:\n", avg_spend))
```
