```python
import json

gradebook = {}

def add_student():
    name = input("Enter student name: ").strip()
    if name in gradebook:
        print("Student already exists.")
        return

    grades_input = input("Enter grades separated by commas (e.g. 88, 92, 75): ").strip()
    try:
        grades = [float(g) for g in grades_input.split(',') if g]
        if not grades:
            raise ValueError
    except ValueError:
        print("Invalid grade input. Please enter number separated by commas.")
        return

    gradebook[name] = grades
    print(f"Added {name} with grades: {grades}")

def remove_student():
    name =  input("Enter name to remove student: ").strip
    if name in gradebook:
        del gradebook[name]
        print(f"removed {name}.")
    else:
        print("Student not found.")


def update_grades():
    name = input("Enter student name to update: ").strip()
    if name not in gradebook:
        print("Student not found.")
        return

    grades_input = input("Enter new grades separated by commas (e.g. 85, 90): ").strip()
    try:
        grades = [float(g) for g in grades_input.split(', ') if g]
        if not grades:
            raise ValueError
        gradebook[name] = grades
        print(f"Updated grades for {name}.")
    except ValueError:
        print("Invalid grade input!")
    


def display_students():
    if not gradebook:
        print("Gradebook is empty.")
        return

    print("\n---Student Gradebook---")
    for name in sorted(gradebook.keys()):
        grades = gradebook[name]
        average = sum(grades) / len(grades) if grades else 0
        print(f"Name: {name}, Grades: {grades}, Average: {average}.2f")

def search_student():
    keyword = input("Enter name to search: ").strip().lower()
    found = False
    for name in gradebook:
        if keyword in name.lower():
            grades = gradebook[name]
            average = sum(grades) / len(grades)
            print(f"Name: {name}, Grades: {grades}, Average: {average}.2f")
            found = True
    if not found:
        print("No matching student found.")
        
def save_gradebook():
    filename = input("Enter filename to save(e.g. gradebook.json): ").strip()
    try:
        with open(filename, 'w') as f:
            json.dump(gradebook, f, indent=4)
        print("Gradebook saved.")
    except Exception as e:
        print(f"Error saving file: {e}")

def load_gradebook():
    global gradebook
    filename = input("Enter filename to load (e.g. gradebook.json): ").strip()
    try:
        with open(filename, 'r') as f:
            gradebook = json.load(f)
        print("Gradebook loaded.")
    except FileNotFoundError:
        print("File not found.")
    except json.JSONDecodeError:
        print("File is not a valid JSON.")
    except Exception as e:
        print(f"An error occured: {e}")
    
def menu():
    while True:
        print("\n--- Student Gradebook Menu ---")
        print("1. Add Student")
        print("2. Remove Student")
        print("3. Search Student")
        print("4. Update Grades")
        print("5. Display All Students")
        print("6. Save Gradebook")
        print("7. Load Gradebook")
        print("8. Exit")

        choice  = input("Choose an option (1-8): ").strip()

        if choice == '1':
            add_student()
        elif choice == '2':
            remove_student()
        elif choice == '3':
            search_student()
        elif choice == '4':
            update_grades()
        elif choice == '5':
            display_students()
        elif choice == '6':
            save_gradebook()
        elif choice == '7':
            load_gradebook()
        elif choice == '8':
            print("Goodbye!")
            break
        else:
            print("Invalid option. Please choose 1-8.")

menu()

    
```
